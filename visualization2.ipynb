```
x=np.linspace(-3,3,100) #-3~3까지 100개로 나눈걸 x에 넣어줌
y1=np.sin(x) #y1 = sin곡선
y2=np.cos(x) #y2 = cos곡선
y3=1/(1+np.exp(-x)) #y3 = log그래프
y4=np.exp(x) #y4 = 지수함수 그래프

fig, ax = plt.subplots(2,2) #(2,2)의 subplot생성

ax[0, 0].plot(x, y1) #x=x, y=y1
ax[0, 1].plot(x, y2)
ax[1, 0].plot(x, y3)
ax[1, 1].plot(x,y4)

ax[0, 0].set_title("Sine function") #제목설정
ax[0, 1].set_title("Cosine function")
ax[1, 0].set_title("Sigmoid function")
ax[1, 1].set_title("Exponential function")

fig.tight_layout() #subplot간 올바른 간격을 자동 유지
```

```
x=np.linspace(-3,3,100)
y1=np.sin(x)
y2=np.cos(x)
y3=1/(1+np.exp(-x))
y4=np.exp(x)

fig, ax = plt.subplots(2, 2)

ax[0, 0].plot(x, y1)
ax[0, 1].plot(x, y2)
ax[1, 0].plot(x, y3)
ax[1, 1].plot(x,y4)

ax[0, 0].set_title("Sine function")
ax[0, 1].set_title("Cosine function")
ax[1, 0].set_title("Sigmoid function")
ax[1, 1].set_title("Exponential function")

plt.subplots_adjust(left=0.125,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, #left,bottom,right,top은 subplot의 4면 위치를 지정
                    wspace=0.2, 
                    hspace=0.35) #subplot 사이에 예약 된 공간을 지정
```

```
im1=np.random.random((50,50)) #im1에 (50,50)크기의 난수를 넣음
im2=np.random.random((40,50))
im3=np.random.random((50,40))
im4=np.random.random((60,50))

plt.subplot(221) #(0,221)subplot생성 
plt.imshow(im1) #im1을 출력
plt.subplot(222)
plt.imshow(im2)
plt.subplot(223)
plt.imshow(im3)
plt.subplot(224)
plt.imshow(im4)

plt.subplot_tool() #위 이미지에 대한 서브 플롯 도구 창을 시작
plt.show()
```

```
a=np.linspace(0,5,100) #0~5까지 100개의 수열생성

#constrained_layout=True로 subplot과 장식 최적의 수치로 자동조정
figure, axes = plt.subplots(2,2, constrained_layout=True)

#Draw a graph
axes[0, 0].plot(x, np.exp(a))
axes[0, 1].plot(a, np.sin(a))
axes[1, 0].plot(a, np.cos(a))
axes[1, 1].plot(range(10)) 

#do title
axes[0, 0].set_title("subplot 1")
axes[0, 1].set_title("subplot 2")
axes[1, 0].set_title("subplot 3")
axes[1, 1].set_title("subplot 4")

plt.show()
```

```
fig = plt.figure(figsize=[12,8]) #그림판의 크기 설정

plt.plot([1,2,3,4], label='up') #1,2,3,4값의 이름이up인 선그래프 생성
plt.plot([4,3,2,1], label='down') #4,3,2,1값의 이름이down인 선그래프 생성
plt.legend(loc='best') #legend를 최적의 위치에 배치
plt.axis('scaled')
plt.axis([-1,4,0,5]) #what is axis참조

plt.title('UP and DOWN lines', fontsize=20) #그래프의 제목을 20폰트크기로 지정
plt.xlabel('X') #x축의 이름을 x로 지정
plt.ylabel('value') #y축의 이름을 value로 지정

plt.xticks([0, 1.5, 3],['one','1 1/2', 'three']) #x축의 0,1.5,3의 이름이 오른쪽list가 됨
plt.yticks(np.arange(1,4.1,0.5)) #y축의 1~4까지 0.5씩 띈곳 마다 이름을 부여
plt.grid()  # 격자출력
```
