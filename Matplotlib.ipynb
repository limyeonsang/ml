수직선/수평선 표시하기

*   hlines = 수평
*   vlines = 수직



```
a = np.arange(0, 2.4, 0.2)

plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)
plt.xticks(np.arange(0, 2.4, 0.2), labels=['Jan', '', 'Mar' , '', 'May', '', 'July', '', 'Sep', '', 'Nov'])
plt.yticks(np.arange(0, 14, 2), ('0', '1GB', '2GB', '3GB', '4GB', '5GB', '6GB'))

plt.axhline(1, 0, 0.45, color='gray', linestyle='--', linewidth='1')
plt.axvline(1, 0, 0.14, color='lightgray', linestyle=':', linewidth='2')
# 1번 인자는 y값으로서 수평선의 위치가 됨
# 두,세번째 인자는 최소, 최대 값으로서 0~1사이임

plt.axhline(10.9, 0, 0.96, color='gray', linestyle='--', linewidth='1')
plt.axvline(2.2, 0, 0.90, color='lightgray', linestyle=':', linewidth='2')

plt.show()
```

```
a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)
plt.xticks(np.arange(0, 2, 0.2), labels=['Jan', '', 'Feb', '', 'Mar', '', 'May', '', 'June', '', 'July'])
plt.yticks(np.arange(0, 7), ('0', '1GB', '2GB', '3GB', '4GB', '5GB', '6GB'))

plt.hlines(4, 1, 1.6, colors='pink', linewidth=3)
# 인자값 순서대로 y, xmin, xmax를 입력하면 점 (xmin,y)에서 점 (xmax,y)를 따라 수평선을 표시
plt.vlines(1, 1, 4, colors='pink', linewidth=3)
# 인자값 순서대로 x, ymin, ymax를 입력하면 점 (x, ymin)에서 점 (x, ymax)를 따라 수평선을 표시

plt.show()
```

산점도

```
np.random.seed(19680801) # 같은 난수를 재사용할 수 있게됨 (0 ~ 4294967295 사이의 값)

N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
# s=마커의 면적, c=마커의 색상, alpha=마커 색의 투명도
plt.show()
```

3d 산점도

```
from mpl_toolkits.mplot3d import Axes3D
```

```
n = 100
xmin, xmax, ymin, ymax, zmin, zmax = 0, 20, 0, 20, 0, 50 # x, y는 0~20 / z는 0~50 사이의 값을 갖도록 범위를 지정
cmin, cmax = 0, 2
xs = np.array([(xmax - xmin) * np.random.random_sample() + xmin for i in range(n)])
ys = np.array([(ymax - ymin) * np.random.random_sample() + ymin for i in range(n)])
zs = np.array([(zmax - zmin) * np.random.random_sample() + zmin for i in range(n)])
color = np.array([(cmax - cmin) * np.random.random_sample() + cmin for i in range(n)])

plt.rcParams['figure.figsize'] = (6,6) # figure의 사이즈 설정
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs, ys, zs, c=color, marker='o', s=15, cmap='Greens')
plt.show()
```

```
a = 2.0 * np.random.randn(10000) + 1.0 # 표준편차 2.0/평균 1.0을 갖는 정규분포
b = np.random.standard_normal(10000) # 표준정규분포를 따름
c = 20.0 * np.random.rand(5000) - 10.0 # -10.0 ~ 10.0사이의 균일한 분포를 같는 값

plt.hist(a, bins=100, density=True, alpha=0.7, histtype='step') # 하늘색
plt.hist(b, bins=50, density=True, alpha=0.5, histtype='stepfilled') # 주황색
plt.hist(c, bins=100, density=True, alpha=0.9, histtype='step') # 연두색

# bins = 몇개의 영역으로 쪼갤지 설정
# density=True = 밀도함수가 되어서 막대 아래 면적이 1이됨
# alpha = 투명도(0.0 ~ 0.1)
# histtype = 'step': 막대 내부가 빔 / 'stepfilled':막대 내부가 참

plt.show()
```

error bar

```
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
yerr = [2.3, 3.1, 1.7, 3.5] # 편차

plt.errorbar(x,y,yerr=yerr)
plt.show()
```

```
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
yerr = [(2.3, 3.1, 1.7, 2.5), (1.1, 2.5, 0.9, 3.9)]
# 비대칭 편차를 표시하기 위해
(2, n)형태의 값을 입력
첫번쨰 튜플은 아래 방향 편차 / 두번쨰 튜플은 위 방향 펴나가 된다.

plt.errorbar(x, y, yerr=yerr)
plt.show()
```

```
x = np.arange(1,5)
y = x**2
yerr = np.linspace(0.1, 0.4, 4)

plt.errorbar(x, y+4, yerr=yerr)
plt.errorbar(x, y+2, yerr=yerr, uplims=True, lolims=True)
# uplims=True/False = True일 때 위쪽 화살표 표시
# lolims=True/False = True일 때 아래쪽 화살표 표시

upperlimits = [True, False, True, False]
lowerlimits = [False, False, True, True]
plt.errorbar(x, y, yerr=yerr, uplims=upperlimits, lolims=lowerlimits)
plt.show()
```

pie chart

```
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']

plt.pie(ratio, labels=labels, autopct='%.1f%%') 
# autopct = 부채꼴 안에 표시될 숫자의 형식

plt.show()
```

```
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False)
# startangle = 부채꼴이 그려지는 시작 각도, default=0 (양의 방향 x축)
# counterclock=False, 시계 방향 순서로 부채꼴 영역이 표시

plt.show()
```

```
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
explode = [0, 0.10, 0, 0.10]

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False,
        explode=explode)
# explode = 부채꼴이 파이 차트의 중심에서 벗어나는 정도

plt.show()
```

```
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
explode = [0.05, 0.05, 0.05, 0.05]

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False,
        explode=explode, shadow=True)
# shadow=True, 그림자 표시

plt.show()
```

```
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
explode = [0.05, 0.05, 0.05, 0.05]
colors = ['silver', 'gold', 'green', 'red']

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, explode=explode, shadow=True, colors=colors)
plt.show()
```
